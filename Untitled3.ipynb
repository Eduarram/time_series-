```python
import datetime as dt
tickets = ['WMT', 'KR', 'NDAQ', 'SHW','GIS', 'HRL', 'JNJ', 'CAG', 'CPB', 'DLTR', 'DG', 'FIVE', 'IMBBY', 'STZ', 'SNY', 'PFE', 'TABAK.PR', 'CL','PG', 'KO']
start_day = "2018-01-01"
end_day = (dt.datetime.now()-dt.timedelta(days=1)).strftime("%Y-%m-%d") # 1 dias atra de la fecha actual

data = yd.download(tickets, start_day, end_day)

data.head()
```

```python
#calculo de los retornos
data = np.log(data)
returns = 100 * data['Adj Close'].pct_change().dropna()
returns.head()
returns.index = pd.to_datetime(returns.index, dayfirst=True)
returns=returns.asfreq('b')
returns =returns.fillna(method='ffill')
```

### optimizacion de la cartera por el metodo de dependencia lineal.

```python
correlation_matrix = returns.corr()
correlation_matrix
```

```python
def portfolio_optimal(matrix_correlations, components):
  from sklearn.decomposition import PCA
  pca=PCA(n_components=components)
  principal_components=pca.fit_transform(correlation_matrix)
  selected_assets = correlation_matrix.columns[pca.components_[0].argsort()[::-1][:components]]
  return selected_assets # return de dictionari with name of columns
```

```python
selected_vars=portfolio_optimal(matrix_correlations=correlation_matrix, components=10)
selected_vars
```

```python
portafolio = returns[selected_vars] # return assets is avariable returned for PCA decomposition
```

```python
log_portafolio = np.log(portafolio)
```

```python
from scipy.stats import shapiro

# Crear una lista para almacenar los resultados de la prueba
results = []

# Iterar a través de las columnas y aplicar la prueba de Shapiro-Wilk
for column in portafolio.columns:
    stat, p_value = shapiro(portafolio[column])
    results.append({'Columna': column, 'Estadístico': stat, 'Valor p': p_value})

# Convertir la lista de resultados en un DataFrame
results_df = pd.DataFrame(results)

# Imprimir los resultados
print(results_df)
```

```python
# prueba de estacionariedad
from statsmodels.tsa.stattools import adfuller

stationary_results = []

for column in portafolio.columns:
  dictio = adfuller(portafolio[column])
  tupla=dictio[4]
  stat, p_value, interval_5 = dictio[0], dictio[1], tupla['5%']
  stationary_results.append({'Columna': column, 'estadistico': stat, 'P valor': p_value, 'confianza al 5%': interval_5})


df_stationary= pd.DataFrame(stationary_results)
print(df_stationary)
```

##### Revisando el AIC.

```python
# Crear un DataFrame para almacenar los resultados
best_orders_df1 = pd.DataFrame(columns=['Activo', 'Best_p', 'Best_q', 'Best_AIC'])

for activo in train_portafolio.columns:
    best_aic = float("inf")  # Inicializa el valor de BIC como infinito positivo
    best_order = (0, 0)  # Inicializa los órdenes p y q

    # Define el rango de órdenes a probar
    p_range = range(1, 10)
    q_range = range(1, 10)

    for p in p_range:
        for q in q_range:
            # Ajusta el modelo GARCH con los órdenes p y q
            modelo_garch = arch_model(train_portafolio[activo], vol="GARCH", p=p, q=q, dist='gaussian')
            resultado_garch = modelo_garch.fit()

            # Calcula el BIC del modelo ajustado
            aic = resultado_garch.aic

            # Si el BIC es menor que el mejor BIC encontrado hasta ahora, actualiza los valores
            if aic < best_aic:
                best_aic = aic
                best_order = (p, q)

    # Agregar el mejor orden y BIC al DataFrame
    best_orders_df1.loc[len(best_orders_df1)] = [activo, best_order[0], best_order[1], best_aic]

# Imprimir el DataFrame con los mejores órdenes para cada columna
best_orders_df1.head(10)
```

#### Revisando el BIC

```python
# Crear un DataFrame para almacenar los resultados
best_orders_df = pd.DataFrame(columns=['Activo', 'Best_p', 'Best_q', 'Best_BIC'])

for activo in train_portafolio.columns:
    best_bic = float("inf")  # Inicializa el valor de BIC como infinito positivo
    best_order = (0, 0)  # Inicializa los órdenes p y q

    # Define el rango de órdenes a probar
    p_range = range(1, 10)
    q_range = range(1, 10)

    for p in p_range:
        for q in q_range:
            # Ajusta el modelo GARCH con los órdenes p y q
            modelo_garch = arch_model(train_portafolio[activo], vol="GARCH", p=p, q=q, dist='gaussian')
            resultado_garch = modelo_garch.fit()
            # Calcula el BIC del modelo ajustado
            bic = resultado_garch.bic

            # Si el BIC es menor que el mejor BIC encontrado hasta ahora, actualiza los valores
            if bic < best_bic:
                best_bic = bic
                best_order = (p, q)

    # Agregar el mejor orden y BIC al DataFrame
    best_orders_df.loc[len(best_orders_df)] = [activo, best_order[0], best_order[1], best_aic]

# Imprimir el DataFrame con los mejores órdenes para cada columna
best_orders_df.head(10)
```

```python
modelos_garch = []

# Recorrer las filas del DataFrame
for index, row in best_orders_df.iterrows():
    activo = row['Activo']
    best_p = int(row['Best_p'])
    best_q = int(row['Best_q'])

    # Ajustar un modelo GARCH con los valores de "Best_p" y "Best_q" para esta fila
    modelo_garch = arch_model(train_portafolio[activo], vol='GARCH', p=best_p, mean='constant', q=best_q, dist='gaussian')
    resultado_garch = modelo_garch.fit()

    # Agregar el modelo ajustado a la lista de modelos
    modelos_garch.append(resultado_garch)

# Crear un DataFrame para almacenar los resultados de los modelos
resultados_modelos = pd.DataFrame()

# Agregar los resultados de los modelos a un DataFrame
for i, row in best_orders_df.iterrows():
    activo = row['Activo']
    resultado_garch = modelos_garch[i]
    #resultados_modelos[activo + '_mu'] = resultado_garch.params['mu']
    resultados_modelos[activo + '_omega'] = resultado_garch.params['omega']
    resultados_modelos[activo + '_alpha[1]'] = resultado_garch.params['alpha[1]']
    resultados_modelos[activo + '_beta[1]'] = resultado_garch.params['beta[1]']

# resultados_modelos ahora contiene los parámetros de los modelos GARCH para cada activo en columnas separadas.
```

```python
#train_retornos = pow(train_retornos, 2)
from tsmoothie.utils_func import sim_randomwalk
from tsmoothie.smoother import LowessSmoother

data = train_retornos.dropna().values.reshape(1, -1)

# operate smoothing
smoother = LowessSmoother(smooth_fraction=0.1, iterations=1)
smoother.smooth(data)

# generate intervals
low, up = smoother.get_intervals('prediction_interval')

points = smoother.data[0]
up_points = up[0]
low_points = low[0]

for i in range(len(points)-1, 0, -1):
    current_point = points[i]
    current_up = up_points[i]
    current_low = low_points[i]
    if current_point > current_up or current_point < current_low:
       print(f'found an outlier value: {current_point}')
```

```python
outlier_dates = []
outlier_values = []

for i in range(len(points)-1, 0, -1):
    current_point = points[i]
    current_up = up_points[i]
    current_low = low_points[i]
    if current_point > current_up or current_point < current_low:
        outlier_dates.append(train_retornos.index[i])
        outlier_values.append(current_point)

outliers_df = pd.DataFrame({'Date': outlier_dates, 'Value': outlier_values})

# Visualizar la serie de tiempo y los valores atípicos
plt.figure(figsize=(12, 6))
plt.plot(train_retornos, label='Serie de Tiempo')
plt.scatter(outliers_df['Date'], outliers_df['Value'], color='red', label='Valores Atípicos')
plt.xlabel('Fecha')
plt.ylabel('Valor')
plt.legend()
plt.title('Serie de Tiempo con Valores Atípicos')
plt.grid(True)
plt.show()
```

#### Remplazar los outliers

```python
# oreder outliers by ascendin_values
outliers_df.sort_values(by='Value', inplace=True, ascending=False)

outlier_sample=5
#chose the outliers most extremes
last_value = len(outliers_df.index)
rango = last_value - outlier_sample

outliers_max = outliers_df.iloc[0:outlier_sample]
outliers_min = outliers_df.iloc[rango:last_value]
outliers_replace = pd.concat([outliers_max,outliers_min], axis=0)

#order by date
outliers_replace.Date = pd.to_datetime(outliers_replace.Date)
outliers_replace.sort_values(by='Date', inplace=True, ascending=True)
outliers_replace.set_index('Date', inplace=True)
```
